Viết chương trình cho phép người dùng nhập chuỗi và và hiển thị số ký tự trong chuỗi vừa nhập vào.

```python
def count(txt):
    k = len(txt)
    return k

txt = input('input string: ')
print('Length of string: ', count(txt))
```

```python
def count(txt):
    k = 0
    for i in txt:
        k += 1
    return k

txt = input('input string: ')
print('Length of string: ', count(txt))
```
